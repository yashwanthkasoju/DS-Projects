categorizing vote_average  column

```python
def categorize_col(df, col, labels):
    desc = df[col].describe()
    edges = [desc['min'], desc['25%'], desc['50%'], desc['75%'], desc['max']]

    # Check for uniqueness in edges
    unique_edges = sorted(set(edges))
    if len(unique_edges) < 2:
        raise ValueError(f"Not enough unique bin edges to create bins: {edges}")

    # Use pd.cut with duplicates='drop' to handle repeated edges gracefully
    df[col] = pd.cut(
        df[col],
        bins=edges,
        labels=labels[:len(edges)-1],  # adjust labels count accordingly
        include_lowest=True,
        duplicates='drop'
    )
    return df
```

```python
df.dropna(inplace=True)
df.isna().sum()
```
